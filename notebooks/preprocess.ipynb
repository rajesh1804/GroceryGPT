```python
import pandas as pd
```

```python
products = pd.read_csv('../data/raw/products.csv')
aisles = pd.read_csv('../data/raw/aisles.csv')
departments = pd.read_csv('../data/raw/departments.csv')
```

```python
# Merge everything
df = products.merge(aisles, on='aisle_id').merge(departments, on='department_id')
df['full_text'] = df['product_name'] + " " + df['aisle'] + " " + df['department']
```

```python
df.to_csv('../data/products_cleaned.csv', index=False)
```
